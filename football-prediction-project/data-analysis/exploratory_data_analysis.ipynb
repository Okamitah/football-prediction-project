## **4. Simulation**

In this section, we'll make quick *simulations*, to see if we can *profit* form this data.

### **a. Simplest interpretation**

As a first step, we'll try to see if, for **Premier League**, **La Liga** and **Ligue 1**, we can *profit* if we bet respectively on Home, Draw and Away, seeing that they're the ones with the most *occurances* of each outcome.

```python
strategies = {
    'Premier League': 1,  # Home Wins
    'La Liga': 0,         # Draws
    'Ligue 1': 2          # Away Wins
}

bet_size = 1
results = []

for league, target_outcome in strategies.items():
    league_df = df[league]
    league_profit = 0
    cumulative_profit = []

    for _, row in league_df.iterrows():
        if row['Outcome'] == target_outcome:
            if target_outcome == 1:
                profit = (row['Home odds'] * bet_size - bet_size)
            elif target_outcome == 0:
                profit = (row['Draw odds'] * bet_size - bet_size)
            elif target_outcome == 2:
                profit = (row['Away odds'] * bet_size - bet_size)
        else:
            profit = -bet_size

        league_profit += profit
        cumulative_profit.append(league_profit)

    results.append({
        'League': league,
        'Total Profit': league_profit,
        'Cumulative Profit': cumulative_profit
    })
```

It's visualisation:

```python
results_df = pd.DataFrame([
    {'League': result['League'], 'Total Profit': result['Total Profit']}
    for result in results
])

plt.figure(figsize=(8, 6))
sns.barplot(x='League', y='Total Profit', data=results_df, palette='Set2')
plt.title('Total Profit by League Strategy')
plt.xlabel('League')
plt.ylabel('Total Profit')
plt.show()
```

And seeing it's evolution through time

```python
plt.figure(figsize=(10, 6))

for result in results:
    plt.plot(result['Cumulative Profit'], label=result['League'])

plt.title('Cumulative Profit Over Matches')
plt.xlabel('Match Index')
plt.ylabel('Cumulative Profit')
plt.legend()
plt.show()
```

We can clearly see that it's not as simple as that. That even when we had a *positive* *profit* in Ligue 1, it was really *small*.
Nonetheless, this observations will come in handy in the next steps.

### **b. Factoring in rating naively**

To make our model better, we'll factor in rating before the match, to get better results, but not in a deep way.

So basically, if home rating - 0.1 > away rating, bet on home, if home rating + 0.1 < away rating, bet on away, and else, on draw.

```python
bet_size = 1
results = []

for league, league_df in df.items():
    league_profit = 0
    cumulative_profit = []

    for _, row in league_df.iterrows():
        if row['Home Last Avg Rating'] - 0.1 > row['Away Last Avg Rating']:
            target_outcome = 1
            odds = row['Home odds']
        elif row['Home Last Avg Rating'] + 0.1 < row['Away Last Avg Rating']:
            target_outcome = 2
            odds = row['Away odds']
        else:
            target_outcome = 0
            odds = row['Draw odds']

        if row['Outcome'] == target_outcome:
            profit = (odds * bet_size - bet_size)
        else:
            profit = -bet_size

        league_profit += profit
        cumulative_profit.append(league_profit)

    results.append({
        'League': league,
        'Total Profit': league_profit,
        'Cumulative Profit': cumulative_profit
    })
```

You know the drill, we have to visualise it

```python
results_df = pd.DataFrame([
    {'League': result['League'], 'Total Profit': result['Total Profit']}
    for result in results
])

plt.figure(figsize=(8, 6))
sns.barplot(x='League', y='Total Profit', data=results_df, palette='Set2')
plt.title('Total Profit by League Strategy (Based on Ratings)')
plt.xlabel('League')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for result in results:
    plt.plot(result['Cumulative Profit'], label=result['League'])

plt.title('Cumulative Profit Over Matches (Based on Ratings)')
plt.xlabel('Match Index')
plt.ylabel('Cumulative Profit')
plt.legend()
plt.show()
```

We ended having *worse* results! I wonder why is that?
It is because we didn't factor in two things: **Home advatange** (a team playing at its home has more chances to win), and the **odds** provided by the bookie.

### **c. Factoring in the odds and home advantage**

At first, we'll just try to apply this logic:

    Bet on Home Wins if:
    • Home Advantage > 0 and  Home Odds > 2, or
    • Home Advantage > −0.05 and  Home Odds > 2.5.

    Bet on Away Wins if:
    •  Home Advantage < −0.2 and  Away Odds > 2, or
    •  Home Advantage < −0.25 and  Away Odds > 2.5.
    
    Bet on Draws in all other cases.

```python
bet_size = 1
results = []

for league, league_df in df.items():
    league_profit = 0
    cumulative_profit = []

    for _, row in league_df.iterrows():
        home_advantage = row['Home Last Avg Rating'] - row['Away Last Avg Rating']

        if home_advantage > 0 and row['Home odds'] > 2:
            target_outcome = 1  # Bet on Home Win
            odds = row['Home odds']
        elif home_advantage > -0.05 and row['Home odds'] > 2.5:
            target_outcome = 1  # Bet on Home Win
            odds = row['Home odds']
        elif home_advantage < -0.2 and row['Away odds'] > 2:
            target_outcome = 2  # Bet on Away Win
            odds = row['Away odds']
        elif home_advantage < -0.25 and row['Away odds'] > 2.5:
            target_outcome = 2  # Bet on Away Win
            odds = row['Away odds']
        else:
            target_outcome = 0  # Bet on Draw
            odds = row['Draw odds']

        if row['Outcome'] == target_outcome:
            profit = (odds * bet_size - bet_size)
        else:
            profit = -bet_size

        league_profit += profit
        cumulative_profit.append(league_profit)

    results.append({
        'League': league,
        'Total Profit': league_profit,
        'Cumulative Profit': cumulative_profit
    })
```

```python
results_df = pd.DataFrame([
    {'League': result['League'], 'Total Profit': result['Total Profit']}
    for result in results
])

plt.figure(figsize=(8, 6))
sns.barplot(x='League', y='Total Profit', data=results_df, palette='Set2')
plt.title('Total Profit by League Strategy (Based on Ratings and Odds)')
plt.xlabel('League')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for result in results:
    plt.plot(result['Cumulative Profit'], label=result['League'])

plt.title('Cumulative Profit Over Matches (Updated Parameters)')
plt.xlabel('Match Index')
plt.ylabel('Cumulative Profit')
plt.legend()
plt.show()
```

We ended up with worse results for Ligue 1 and Premier League, but better for Bundesliga and Serie A. We can conclude that each league has its *style of play* that influences *outcomes*.
